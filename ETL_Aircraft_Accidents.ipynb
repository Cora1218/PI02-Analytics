```python
# Dataset loading.
df = pd.read_csv('Dataset/AccidentesAviones.csv')
df.info() # Dataset information.
```

```python
# Count the missing values of each column.
value_count = df['Airline'].value_counts()
# Show the complete table of results.
pd.set_option('display.max_rows', None)  # Show all the rows.

print(value_count)
```

```python
# The following columns were eliminated as they do not contribute significant value: HORA declarada, flight_no, registration, cn_ln, PASAJEROS A BORDO, crew_aboard, passenger_fatalities, crew_fatalities, ground.
columnas_eliminar = ['HORA declarada', 'flight_no', 'registration', 'cn_ln', 'PASAJEROS A BORDO', 'crew_aboard', 'passenger_fatalities', 'crew_fatalities', 'ground']
df = df.drop(columnas_eliminar, axis=1)
df.info()
```

```python
# Rename columns to standardize.
new_names = ['Accident_Date', 'Accident_Route', 'Airline', 'Departure_Destination', 'Aircraft_Type', 'People_Aboard', 'Number_Fatalities', 'Summary']
df.columns = new_names
df.info()
df.head()
```

Impute missing data '?' by mean or median value without affecting data integrity because the missing values '?' are few.
# I decided to replace with the median because the result was smaller than the mean.

```python
# Calculate the mean or the median of the column 'People_Aboard' y 'Number_Fatalities' without missing '?' values.
# value_mean_people = df['People_Aboard'][df['People_Aboard'] != '?'].astype(int).mean()
value_median_people = df['People_Aboard'][df['People_Aboard'] != '?'].astype(int).median()

#value_mean_fatalities = df['Number_Fatalities'][df['Number_Fatalities'] != '?'].astype(int).mean()
value_median_fatalities = df['Number_Fatalities'][df['Number_Fatalities'] != '?'].astype(int).median()

# Replace the missing values '?' with the mean or the median.
df['People_Aboard'] = df['People_Aboard'].replace('?', int(value_median_people))
df['Number_Fatalities'] = df['Number_Fatalities'].replace('?', int(value_median_fatalities))

df.info()
```

Change the appropriate format for each column.

```python
# Change the data type of the column 'People_Aboard' and 'Number_Fatalities'' from object to int.
df['People_Aboard'] = df['People_Aboard'].astype(int)
df['Number_Fatalities'] = df['Number_Fatalities'].astype(int)

# Change the data type of the column 'Accident_Date' from object to date.
df['Accident_Date'] = pd.to_datetime(df['Accident_Date'])

df.info()
```
